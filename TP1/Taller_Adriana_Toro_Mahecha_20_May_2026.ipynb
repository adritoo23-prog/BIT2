Parte 2 — Limpieza de datos (Pandas)


```python
##No se evidencian registros duplicados.
df.duplicated().sum()
```

```python
#Corregir tipos de datos incorrectos; sin embargo se genra una copia de DF para hacer la limpieza

dfcopy1= df.copy()
```

Se valida la información de cada columna, con el fin de evidenciar datos a corregir y unificar

```python
dfcopy1["cliente"].unique()

#nada por corregir
```

```python
dfcopy1["producto"].unique()

#nada por corregir
```

```python
dfcopy1["fecha"].unique()

#nada por corregir pero si cambiar a tipo de dato fecha
```

```python
#fecha

dfcopy1["fecha"]=pd.to_datetime(dfcopy1["fecha"])
dfcopy1["fecha"].unique()
```

```python
dfcopy1.info()
```

```python
dfcopy1["pais"].unique()

#se debe corregir y unificar datos en la columna
```

```python
#unificar Mayusculas con Title
dfcopy1["pais"]=dfcopy1["pais"].str.title()
dfcopy1["pais"].unique()
```

```python
#unificar pais
dfcopy1["pais"]=dfcopy1["pais"].replace({
    "Perú":"Peru",
    "Col":"Colombia"
})
dfcopy1["pais"].unique()
```

```python
dfcopy1["cantidad"].unique()

#unificar a numero la información
```

```python
#se corrige texto a número

dfcopy1["cantidad"]=dfcopy1["cantidad"].replace({
    "three":"3"
})
dfcopy1["cantidad"].unique()
```

```python
#valores nulos identificados, convertir columna a número

dfcopy1["cantidad"]=pd.to_numeric(dfcopy1["cantidad"], errors="coerce")
dfcopy1["cantidad"].unique()
```

```python
#nulos

dfcopy1["cantidad"]=dfcopy1["cantidad"].fillna(dfcopy1["cantidad"].median())
dfcopy1["cantidad"].unique()
```

```python
dfcopy1["metodo_pago"].unique()

#unificar metodos de pago
```

```python
dfcopy1["metodo_pago"]=dfcopy1["metodo_pago"].str.title()
dfcopy1["metodo_pago"].unique()
```

```python
#manejo outliers

dfcopy1= dfcopy1[dfcopy1["precio"]<5000]
dfcopy1.describe(include="all")
```
